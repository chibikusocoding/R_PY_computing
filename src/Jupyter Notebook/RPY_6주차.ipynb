```python
hand = open('mbox-short.txt')
emails=[]
for line in hand:
    line = line.rstrip() #오른쪽 공백열 제거 
    if re.search('^From:.+@', line): # 괄호없이 ^이면 처음부터 매치 되는지 확인, 이메일 쓴사람을 찾아보자  
        emails.append(line) # 데이터화 
emails
```

```python
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('[a-zA-Z0-9]\S+@\S+[a-zA-Z]', line) # 이메일 주소 아무거나 찾아보자라는 느낌 
    if len(x)>0:
      print(x)
```

```python
hand = open('mbox-short.txt')
probs=[]
for line in hand:
    line = line.rstrip()
    x = re.findall('^X\S*: ([0-9.]+)', line) # 소괄호는 Extraction 
    if len(x) > 0:
        probs.extend(x)
probs
hand = open('mbox-short.txt')
contents=[]
for line in hand:
    line = line.rstrip()
    x = re.findall('(^X\S*): [0-9.]+', line) # 소괄호는 Extraction 
    if len(x) > 0:
        contents.extend(x)
contents

import pandas as pd

pd.DataFrame({"prob":probs,"content":contents })
# 궁국적인 목표는 DB화 : 데이터 사이언스의 기초! 
```

```python
boston_url = 'https://andybek.com/pandas-marathon'
boston = pd.read_csv(boston_url)
boston.info()
# Pandas Text 명령어 
boston.City.str.upper() 
boston.City.str.lower()
boston.City.str.swapcase()
boston.City.str.capitalize()
boston.Name.iloc[0:2].str.strip()
boston.Name.str.split(',')
boston.Name.str.split(',')[0]
boston.Name.str.split(',').str.get(0)
boston['First Name'] = boston.Name.str.split(',').str.get(1) # 두번째것 
boston['Last Name'] = boston.Name.str.split(',').str.get(0) # 첫번째것 
boston.head()
boston['M/F'].str.cat(boston.Age.astype(str), sep='_') # M/F의 문자열을 뽑아서 Age 변수를 문자열로 바꾼후 _를 붙여서 합친다
```

```python
boston.Country.str.slice(0, 2, 1)
boston.Country.str.slice(start=-2, step=1, stop=None)
# regular expression 사용 가능한 환경 
boston.Country.str.match('ITA')
boston[boston.Country.str.match('ITA')]
wills =  boston.Name.str.contains('Will')
print(wills)
```

```python
boston.Name[boston.Name.str.contains('will')] # 대문자만 찾는것 
boston.Name[boston.Name.str.contains('[wW]ill', regex=True)] #정규 표현식을 통해 필터 강화 
boston.Name[boston.Name.str.contains(',\s[wW]ill$', regex=True)]
boston.Name[boston.Name.str.contains(',\s[wW]ill\b', regex=True)]
boston['Official Time'].head()
boston['Official Time'].str.replace('(\d+):(\d+):(\d+)', r'\1 hours, \2 minutes, and \3 seconds') #정규 표현식 패턴 매치 
```

```python

# parse xml file
import xml.etree.ElementTree as ET
tree = ET.parse('country_data.xml') # XML 데이터를 파이썬이 이해하는 형식으로 바꿈 
root = tree.getroot()
print(root.tag)
print(root.attrib)
for child in root:
	print(child.tag, child.attrib)

root[0][1].tag
root[0][1].text
```

```python
for neighbor in root.iter('neighbor'): # neihbor라는 태그를 모두 찾아서 
	print(neighbor.attrib)

for country in root.findall('country'):
	rank = country.find('rank').text # 특정 태그의 텍스트를  뽑자 
	name = country.get('name') # 특정 태그의 속성값의 텍스트를 뽑자 
	print (name, rank)
```
