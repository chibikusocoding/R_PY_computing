```python
#Drill 401
# some python lists
names = ['Olga', 'Andrew', 'Brian', 'Telulah', 'Nicole', 'Tilda']
ages = [29, 21, 45, 23, 39, 46]
married = [False, True, True, True, False, True]

# pandas series -> 한개열로 데이터 만드는 것 
ser = pd.Series(names, name='name')
print(ser)
#pandas dataframe-> 일종의 엑셀 테이블 만들기 
df = pd.DataFrame({'name': names, 'age': ages, 'married': married}) # 딕셔너리 쓰는 법 
print(df)

#iloc의 사용법 
ser.iloc[2] # Series의 세번째 요소를 불러오겠다 
df.iloc[2,0] # Dataframe에 3번째 행, 1번째 열  
# Shape 메소드: 데이터 실제 차원  
ser.shape
df.shape
df.dtypes
df.info()
df.info(verbose=True)
```

```python
# Drill 402 
dataurl = 'https://andybek.com/pandas-nutrition'
nutrition = pd.read_csv(dataurl) # CSV 파일로 읽자 
nutrition.head(10)
nutrition.info(verbose=False, memory_usage='deep')
nutrition.head()
nutrition['Unnamed: 0'] # 이해해보자->unamed 0이라는 칼럼을 추출하자 
#nutrition.drop('Unnamed: 0', axis=1) # 칼럼삭제, 열축에서 
nutrition.set_index('Unnamed: 0')

# 인덱스 조정 
nutrition = pd.read_csv(dataurl, index_col=[0]) # 첫행은 인덱스 행이다로 지정 
nutrition.iloc[2,0]
#Sample 메소드 : 관측치 샘플링 
nutrition.sample()
nutrition.sample(random_state=12)
nutrition.sample(n=3)
nutrition.sample(frac=0.01)
```

```python
nutrition = pd.read_csv(dataurl, index_col=[0]) # 첫행은 인덱스 행이다로 지정 

# Axes 메소드
print(nutrition.axes) # 행이름, 열이름 파악 
print(nutrition.axes[0][3]) # 행이름중 세번째
print(nutrition.index[3]) # 행이름중 세번째 
print(nutrition.axes[1][69]) # 아래와 동일 
print(nutrition.columns[69]) # 열이름중 69번째 


# dropna 메소드
print(nutrition.dropna(axis=0)) # 관측치중 NA 제외 (행에서 axis=0 )

# Index 변화
print(nutrition.index)
print(nutrition.set_index('name', inplace=True))
```

```python
nutrition
```

```python
# LOC 메소드: 인덱스/칼럼 이름기반 호출법 
nutrition.head
nutrition.loc['Eggplant, raw'] # 행우선 
nutrition.loc['Eggplant, raw']['calories'] # 행/열 부르는 법 
nutrition.loc['Eggplant, raw', 'calories']
nutrition.loc['Eggplant, raw':'Sherbet, orange', 'calories':'cholesterol']
nutrition.loc[
              ['Raspberries, raw', 'Blackberries, raw'],
              ['protein', 'vitamin_b6', 'water']
] # 특정행/특정열 
```

```python
# ILOC 메소드
nutrition.iloc[3] # 4번째 행
nutrition.iloc[3, :]
nutrition.iloc[[4,6,9], :] #, 5, 7,10 행 
nutrition.iloc[[4,6,9]]
nutrition.iloc[[4,6,9], 2] # 5,7,10행 3열
nutrition.iloc[[4,6,9], 2:5] # 5,7,10행 3,4,5열 

# List Creation 법 
[True if i%2==0 else False for i in range(8789)] 
# i가 0부터 8788까지 변하는데 
# i를 2로 나누었을때 나머지가 0이면 True 
# 나머지가 0이 아닌 경우는 False why? ELSE 
# List로 만들어라 
new_nutr = nutrition.iloc[
               [True if i%2==0 else False for i in range(8789)],
               [True if i%2==0 else False for i in range(75)]
]

# Data 선택의 문제
# Loc을 쓰면: 이름기반 행이름(index), 열이름(column)
# ILoc을 쓰면(1): 이름기반 행의 수열, 열의 수열 기반으로 추출한다. 
# ILoc을 쓰면(2): True/False 기반 추출-> True만 뽑아라 
# 2번째의 전제조건은 -> True/False List의 원소개수가 행/열의 갯수와 동일해야 한다. 
#일반적인 방법론
nutrition.loc['Nuts, pecans', 'calories']
nutrition.iloc[1,1]
# 특정원소 뽑을때 
nutrition.at['Nuts, pecans', 'calories']
nutrition.iat[1,1]

# at/iat이 빠르다. 
%timeit nutrition.loc['Nuts, pecans', 'calories']
%timeit nutrition.at['Nuts, pecans', 'calories']












```
