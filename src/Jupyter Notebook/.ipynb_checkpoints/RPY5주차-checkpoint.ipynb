```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'age': [12, 13, 14, 16], 
    'weight': [41.1, 34.5, 83.2, 90.1], 
    'height': ['1.72', '1.74', '1.91', '1.54']
    })
df.info()
df = df.astype(float)
df.astype({'age': int})


```

```python
df.loc[2, 'weight'] = np.nan
df.loc[1, :] = np.nan
df
```

```python
df.dropna()
df.dropna(how='any', axis=0) # 하나라도 NA면 삭제
df.dropna(how='all', axis=0) # 모두다 NA 면 삭제 
df.dropna(thresh=1, axis=0) # 관측치 중 몇개가 NA가 아니냐?
df.dropna(how='any', axis=0)
df.dropna(thresh=df.shape[1], axis=0)
df
df.dropna(axis=1) # 축변환 
df.dropna(how='any', axis=1) # 하나라도 있음 삭제 
df.dropna(how='all', axis=1) # 모두다 NA 삭제 
df.dropna(axis=1, thresh=3)
print(df) # Inplace를 True라고 하지 않으면 실제 바뀌지 않는다
df.dropna(axis=1, thresh=3, inplace=True)
print(df) # Inplace를 True라고 할경우만 df 바뀐다. 
```

```python
# 정렬 기법 배우기 
df = pd.DataFrame({
    'age': [12, 13, 14, 16], 
    'weight': [41.1, 41.1, 41.1, 90.1], 
    'height': ['1.72', '1.74', '1.91', '1.84']
    })
df.info()
df = df.astype(float)
df.astype({'age': int})

# 값의 크기에 따라서 정렬하기 
df.sort_values(by=['weight', 'height'], ascending=[False, True])
```

```python
# 최대 최소 구하기 
df.max(axis=0)
df.min(axis=1)
```

```python
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.info(verbose=True, memory_usage='deep')
players.dtypes.value_counts()
players.market_value > 40 # Players 데이터 프레임에 Market Value가 40보다 크다 작다 판별 
players[players.market_value > 40] # 행에서 Filter하기 
players[players.market_value > 40].shape # 총 13명에대서 선택되고, 17개 열을 보존 
```

```python
players.position.unique()
players.position.unique().size
players.position.isin(['LB', 'CB', 'RB']) # 축약형 명령어 
# 원래는 의미상 
#players.position=="LB" or players.position=="CB" or players.position=="RB" 
players.loc[players.position.isin(['LB', 'CB', 'RB'])]
# True False를 직접 사용해서 선택 

players.market_value.between(40,50)
players[players.market_value.between(40,50)]
players.age <= 25
players.loc[players.age <= 25]
```

```python
#일반적으로 파이썬 조건문은 and/or Pandas dataframe Series간 조건문 결합에는  &/|을 쓴다. 
(players.position=="LB") | (players.position=="CB")
```

```python
players[
        (players.position == 'LB') & 
        (players.age <= 25)
        ]

players[
        (players.position == 'LB') & 
        (players.age <= 25) & 
        (players.market_value >= 10)
        ]
players[
        (players.position == 'LB') &
        (players.age <= 25) &
        (players.market_value >= 10) &
        ~(players.club.isin(['Tottenham', 'Arsenal']))
        ]
```

```python
# 조건식 걸기-2D 
chelsea_23under = (players.club == 'Chelsea') & (players.age.le(23))
players.loc[chelsea_23under, ['position', 'market_value']]
p_cols = players.columns.str.startswith('p')
players.loc[chelsea_23under, p_cols]
players[chelsea_23under]['position']
players.loc[chelsea_23under, 'position']
players.loc[[0, 132], ('name', 'market_value')]
```

```python
# 중복 관측치 제거  
players.duplicated()
players[players.duplicated()]
players.loc[players.duplicated(subset=['club', 'age', 'position', 'market_value'])]
players.loc[players.duplicated(subset=['club', 'age', 'position', 'market_value'])].shape
players.loc[players.duplicated(subset=['club', 'age', 'position', 'market_value'], keep='last')]
players.loc[players.duplicated(subset=['club', 'age', 'position', 'market_value'], keep=False)]
players.drop_duplicates(keep='first') # drop 하고, 처음것만 살린다. 
```

```python
#관측치 제거 
players.drop(labels=19, axis=0) #상하 동일
players.drop(index=19) # 상하동일 
players.drop(index=[19, 20, 21, 231, 10]) # 특정인덱스 지정 
players.drop(labels=['age', 'market_value'], axis='columns')
players.drop(columns=['age', 'market_value', 'name'])
```

```python
# Aggregate 함수  
players[["age","position_cat"]].agg('mean')
players.new_signing.mean()
players[["age","position_cat"]].agg(np.min)
players.select_dtypes(np.number).agg(np.min) # 숫자인 애들만 뽑아서 
players.select_dtypes(np.number).agg(['min', 'max', 'mean'])
players.loc[:, ['market_value', 'fpl_value']].transform(lambda x: x * 0.91) # 함수 적용하여 변환 
players.loc[:, ['market_value', 'fpl_value']] * 0.91

```

```python
# 칼럼에 대해서 특정함수 적용 입니다. 
def round_floats(x):
  if x.dtype == np.float64:
    return round(x)
  return x


players.apply(round_floats)
players.select_dtypes(np.float64).head()
players.select_dtypes(np.float64).apply(round_floats).head()
```

```python
# Demo of variance
high_var_array = np.array([1, 100, 200, 300, 4000, 5000])
low_var_array = np.array([2, 4, 6, 8, 10])
np.var(high_var_array), np.var(low_var_array)
np.std(high_var_array), np.std(low_var_array)
# The standard deviation is the square root of the variance
np.sqrt(np.var(high_var_array))

```
